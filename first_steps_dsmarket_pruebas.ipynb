```python
# Queremos mostrar las ventas totales por ciudad y mostrarlo con un barplot, para ello transformaremos el dataframe pd_sales

pd_sales['total_por_fila'] = pd_sales.iloc[:,7:].sum(axis=1)
df_barplot_total = pd_sales.groupby('region')['total_por_fila'].sum().reset_index()
df_barplot_total
```

```python
sns.barplot(data=df_barplot_total, x = 'region', y = 'total_por_fila')
```

```python
df_barplot_total_2 = pd_sales.groupby('store_code')['total_por_fila'].sum().reset_index()
df_barplot_total_2
```

```python
# a continuacion mostramos lo mismo pero con el detalle de cada tienda
sns.barplot(data=df_barplot_total_2, x = 'store_code', y = 'total_por_fila')
plt.show()
```

```python
df_barplot_total_3 = pd_sales.groupby(['region','store_code'])['total_por_fila'].sum().reset_index()
df_barplot_total_3
```

```python
fig, ax = plt.subplots(figsize=(16,8))

sns.barplot(data=df_barplot_total_3, x = 'region', y = 'total_por_fila', hue = 'store_code')
plt.show()
```

```python
df_barplot_total_4 = pd_sales.groupby(['region','category'])['total_por_fila'].sum().reset_index()
df_barplot_total_4
```

```python
fig, ax = plt.subplots(figsize=(16,8))

sns.barplot(data=df_barplot_total_4, x = 'region', y = 'total_por_fila', hue = 'category')
plt.show()
```
